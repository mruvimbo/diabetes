## Load data

```python
# Load Dataset

data_path = '/content/drive/MyDrive/Colab Notebooks/diabetes.xlsx'
data = pd.read_excel(data_path)
data
```

##Split the data to x and y

```python
y = data ['Diabetes Status']
y
```

```python
x = data.drop('Diabetes Status', axis = 1)
x
```

## Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
x_test, x_train, y_test, y_train = train_test_split(x,y, test_size= 0.2, random_state  = 100 )
```

## Build the model

```python
from sklearn.linear_model import LinearRegression

diabetes_lr = LinearRegression()

diabetes_lr.fit(x_train, y_train)
```

## Apply model to make prediction

```python
diabetes_lr_train_pred = diabetes_lr.predict(x_train)
diabetes_lr_test_pred = diabetes_lr.predict(x_test)

```

```python
diabetes_lr_train_pred
```

```python
diabetes_lr_test_pred
```

## Evaluating the model

```python
## Evaluating the model

from sklearn.metrics import mean_squared_error, r2_score

diabetes_lr_train_mse = mean_squared_error(y_train,diabetes_lr_train_pred)
diabetes_lr_train_r2 = r2_score(y_train,diabetes_lr_train_pred)

diabetes_lr_test_mse = mean_squared_error(y_test,diabetes_lr_test_pred)
diabetes_lr_test_r2 = r2_score(y_test,diabetes_lr_test_pred)
```

```python
## mean square value explains the difference between the predicted value and the actual value
print('diabetes_lr_train_mse',diabetes_lr_train_mse)
print('diabetes_lr_test_mse',diabetes_lr_test_r2)
## *********************interpretation**************************
## The smaller the Mean Square Value the better model performance
## The higher the Mean Square Value the larger the prediction error

## the R2 score measures how well a regression model fits the data
print('diabetes_lr_train_r2', diabetes_lr_train_r2)
print('diabetes_lr_test_r2', diabetes_lr_test_r2)
## *******************interpretation********************************
## if R2 is close to 1 the model is a good fit
## if R2 is close to 0 the poor the fit
## if r2 is close <0 the worst the model
```

```python
## Improving the layout by creating a dataframe

diabetes_lr_results = pd.DataFrame (['Liner Regression', diabetes_lr_train_mse, diabetes_lr_train_r2,diabetes_lr_test_mse,diabetes_lr_test_r2]).transpose()
diabetes_lr_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE','Test R2']
diabetes_lr_results
```

## Which is the best model?

Based on the provided results, the **Random Forest** model is the best choice. Here's the reasoning:

1. **Performance on Test Data (Most Important Criterion)**:
   - Random Forest has the **lowest Test MSE (0.168745)** and the **highest Test R² (0.258422)** among the three models, indicating it generalizes better to unseen data compared to the others.
   - Linear Regression performs similarly but slightly worse on the test set (higher MSE and lower R²).
   - Decision Tree is clearly **overfitting**, as it has a perfect Train R² (1.0) but terrible Test R² (-0.703), making it unusable for real-world predictions.

2. **Balance Between Bias and Variance**:
   - Random Forest shows a small gap between Train MSE (0.149) and Test MSE (0.168), suggesting it generalizes well without severe overfitting.
   - Linear Regression has consistent but mediocre performance (high bias), while Decision Tree fails entirely due to overfitting (high variance).

3. **Practical Consideration**:
   - Though Random Forest's improvement over Linear Regression is modest, it is still the best-performing model overall. If interpretability is critical, Linear Regression could be considered, but Random Forest is superior for predictive accuracy.

### Conclusion:
**Random Forest** is the best model because it achieves the best test performance while avoiding overfitting. The Decision Tree is disqualified due to extreme overfitting, and Linear Regression is outperformed by Random Forest on both MSE and R² for the test set.
